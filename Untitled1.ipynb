Pour comparer l'algorithme du gradient stochastique (SGD), le gradient de descente classique (Batch Gradient Descent) et le mini-batch gradient descent en utilisant un réseau de neurones simple, nous utiliserons Python avec la bibliothèque TensorFlow et Keras pour illustrer cette comparaison.
L'objectif de ce modèle est de classifier des images de chiffres manuscrits provenant du dataset MNIST  ce modèle de réseau de neurones doit apprendre à reconnaître et à prédire quel chiffre (de 0 à 9) est représenté dans une image donnée.
Nous allons ensuite comparer les performances et le temps de convergence du Batch Gradient Descent, du Stochastic Gradient Descent, et du Mini-batch Gradient Descent.

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
import matplotlib.pyplot as plt
import numpy as np
```

```python
import numpy as np

def metropolis_hastings_pi(N, delta):
    # Initialisation
    x, y = 0.0, 0.0
    count_inside_circle = 0

    for i in range(N):
        # Générer un nouveau point candidat
        x_new = x + delta * (np.random.uniform(-1, 1))
        y_new = y + delta * (np.random.uniform(-1, 1))

        # Vérifier si le nouveau point est dans le carré unité
        if abs(x_new) <= 1 and abs(y_new) <= 1:
            # Probabilité d'acceptation
            if x_new**2 + y_new**2 <= 1:
                alpha = 1.0
            else:
                alpha = 0.0

            u = np.random.uniform(0, 1)
            if u < alpha:
                x, y = x_new, y_new

        # Compter les points dans le cercle
        if x**2 + y**2 <= 1:
            count_inside_circle += 1

    # Estimer pi
    pi_estimate = 4 * (count_inside_circle / N)
    return pi_estimate

# Paramètres de l'algorithme
N = 1000000  # Nombre d'itérations
delta = 0.1  # Longueur de pas

# Estimation de pi
pi_estimated = metropolis_hastings_pi(N, delta)
print(f"Estimation de pi: {pi_estimated}")
```

```python
# Charger le dataset MNIST
mnist = tf.keras.datasets.mnist  #est une référence au module MNIST dans la bibliothèque Keras, qui est incluse dans TensorFlow. Ce module contient des fonctions pour charger le dataset MNIST.
(x_train, y_train), (x_test, y_test) = mnist.load_data() #(x_train, y_train) : Les images et les labels (étiquettes) pour l'entraînement.
                                                           #(x_test, y_test) : Les images et les labels pour le test
x_train, x_test = x_train / 255.0, x_test / 255.0  # Normalisation des données

# Remodeler les données
x_train = x_train.reshape(-1, 784)
x_test = x_test.reshape(-1, 784)
```

    Les images dans x_train et x_test sont initialement représentées par des valeurs de pixels allant de 0 à 255.
    La normalisation est effectuée en divisant chaque valeur de pixel par 255.0, ce qui transforme les valeurs des pixels pour qu'elles se situent dans l'intervalle [0, 1].
    La normalisation aide les réseaux de neurones à converger plus rapidement et à améliorer les performances de l'entraînement.

    Remodeler les données
        reshape(-1, 784) transforme chaque image de 28x28 pixels en un vecteur de 784 éléments (28*28 = 784).
    Le -1 permet à NumPy ou TensorFlow de déterminer automatiquement le nombre de samples basé sur la dimension totale et les autres dimensions spécifiées. Par exemple, si x_train contient 60 000 images de 28x28, le reshape va transformer x_train en un tableau de dimension (60000, 784).
    Cette transformation est nécessaire parce que la plupart des modèles de réseaux de neurones denses (Fully Connected) attendent des vecteurs en entrée plutôt que des matrices (images).

```python
#construit et retourne un modèle de réseau de neurones en utilisant l'API séquentielle de Keras.
def create_model():
    model = Sequential([
        Dense(128, activation='relu', input_shape=(784,)), # premiére couche de 128 neuronne
        Dense(10, activation='softmax')# deuxiéme couche de  10 neurones qui correspond au nombre de classes dans le dataset MNIST (les chiffres de 0 à 9).
    ])
    return model
```

    Sequential est une classe de Keras qui est utilisée pour empiler des couches de manière linéaire.
    La création d'un objet Sequential signifie que nous allons ajouter des couches à ce modèle, une par une, dans l'ordre.

```python
# Méthode 1 : Batch Gradient Descent

batch_model = create_model()
#Configure le modèle avec l'optimiseur Stochastic Gradient Descent (SGD), une fonction de perte adaptée à la classification multi-classes
batch_model.compile(optimizer=SGD(learning_rate=0.01), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
#Entraîne le modèle sur 10 époques
batch_history = batch_model.fit(x_train, y_train, epochs=10, batch_size=len(x_train), validation_data=(x_test, y_test), verbose=0)

# Méthode 2 : Stochastic Gradient Descent
sgd_model = create_model()
sgd_model.compile(optimizer=SGD(learning_rate=0.01), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
sgd_history = sgd_model.fit(x_train, y_train, epochs=10, batch_size=1, validation_data=(x_test, y_test), verbose=0)

# Méthode 3 : Mini-batch Gradient Descent
mini_batch_model = create_model()
mini_batch_model.compile(optimizer=SGD(learning_rate=0.01), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
mini_batch_history = mini_batch_model.fit(x_train, y_train, epochs=10, batch_size=32, validation_data=(x_test, y_test), verbose=0)
```

```python
def plot_history(histories, titles):
    plt.figure(figsize=(12, 8))
    for i, history in enumerate(histories):
        plt.plot(history.history['loss'], label=f'{titles[i]} - train')
        plt.plot(history.history['val_loss'], label=f'{titles[i]} - val')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.title('Comparison of Training Methods')
    plt.show()

# Plot the results
plot_history([batch_history, sgd_history, mini_batch_history], ['Batch Gradient Descent', 'Stochastic Gradient Descent', 'Mini-batch Gradient Descent'])
```

Analyse des Résultats
En regardant le graphique des courbes de perte pour les trois méthodes de descente de gradient, nous pouvons tirer plusieurs conclusions :
    1. Batch Gradient Descent :
        ◦ La courbe de perte pour le Batch Gradient Descent (ligne bleue pour l'entraînement et ligne orange pour la validation) montre une diminution très lente de la perte.
        ◦ La perte commence élevée et diminue légèrement, ce qui suggère que cette méthode n'est pas très efficace dans ce contexte en termes de rapidité de convergence.
        ◦ Le Batch Gradient Descent effectue une mise à jour des poids par époque, utilisant l'ensemble complet des données à chaque itération, ce qui peut expliquer pourquoi il semble moins performant dans cette configuration.
    2. Stochastic Gradient Descent (SGD) :
        ◦ Les courbes de perte pour le SGD (ligne verte pour l'entraînement et ligne rouge pour la validation) montrent une perte relativement stable et basse dès le début.
        ◦ Le SGD met à jour les poids après chaque exemple de formation, ce qui permet une convergence rapide mais peut également introduire des fluctuations.
        ◦ Dans ce cas, les courbes de perte sont relativement plates, suggérant que le SGD converge rapidement vers un minimum.
    3. Mini-batch Gradient Descent :
        ◦ Les courbes de perte pour le Mini-batch Gradient Descent (ligne violette pour l'entraînement et ligne marron pour la validation) montrent une diminution rapide de la perte au début, puis une stabilisation.
        ◦ Cette méthode combine les avantages des deux autres, avec des mises à jour plus fréquentes que le Batch Gradient Descent mais moins bruitées que le SGD.
        ◦ Les courbes montrent une bonne performance en termes de réduction de la perte et de stabilité.
Conclusions
    1. Batch Gradient Descent :
        ◦ Lent à converger : Comme la perte diminue très lentement, le Batch Gradient Descent est inefficace pour des tâches nécessitant une convergence rapide.
        ◦ Mémoire et calcul intensif : Utiliser l'ensemble complet des données pour chaque mise à jour est coûteux en mémoire et en calcul, surtout pour de grands datasets.
    2. Stochastic Gradient Descent (SGD) :
        ◦ Convergence rapide : Le SGD montre une convergence rapide, ce qui est un avantage pour des problèmes nécessitant une solution rapide.
        ◦ Fluctuations possibles : Bien que la perte soit stable dans ce cas, le SGD peut introduire des fluctuations importantes dans d'autres contextes.
    3. Mini-batch Gradient Descent :
        ◦ Équilibre entre vitesse et stabilité : Cette méthode offre un bon compromis entre la rapidité de convergence et la stabilité des mises à jour.
        ◦ Efficace en pratique : Pour la plupart des problèmes pratiques, le Mini-batch Gradient Descent est souvent préféré en raison de sa performance équilibrée.
Recommandations
    • Mini-batch Gradient Descent : Recommandé pour la plupart des scénarios en raison de son équilibre entre rapidité et stabilité. Ajustez la taille des mini-batchs pour optimiser les performances en fonction des ressources disponibles.
    • Stochastic Gradient Descent : Peut être utile lorsque des solutions rapides sont nécessaires, mais attention aux fluctuations potentielles. Considérez des techniques comme la moyenne mobile des poids pour lisser les mises à jour.
    • Batch Gradient Descent : Moins pratique pour les grands datasets en raison de sa lenteur et de sa consommation de ressources. Utilisable pour de petits datasets ou pour des démonstrations pédagogiques.
En conclusion, pour des problèmes réels et complexes, le Mini-batch Gradient Descent est souvent la meilleure option, offrant un bon compromis entre rapidité de convergence et stabilité des résultats.


Difficultés rencontrées

Non-convexité de la fonction de perte : La fonction de perte des réseaux de neurones est généralement non-convexe et peut contenir de nombreux minima locaux et points de selle, ce qui complique l'optimisation. Le SGD, en raison de sa nature stochastique, peut aider à éviter les minima locaux en introduisant du bruit dans le processus de mise à jour des poids, ce qui permet au modèle de s'échapper des vallées peu profondes.Dans notre cas on a utiliser : 'sparse_categorical_crossentropy cette fonction est utilisée pour les problèmes de classification multi-classes où les étiquettes sont des entiers (par opposition à categorical_crossentropy où les étiquettes sont encodées en one-hot).
Pendant l'entraînement, cette fonction de perte calcule la différence entre les prédictions du modèle et les étiquettes réelles et retourne un score indiquant combien le modèle se trompe.

Choix du taux d'apprentissage (learning_rate=0.01): Un taux d'apprentissage trop élevé peut entraîner une divergence de l'algorithme, tandis qu'un taux trop bas peut ralentir la convergence. Des techniques comme le taux d'apprentissage adaptatif (ex. Adam, RMSprop) ajustent dynamiquement le taux d'apprentissage pendant l'entraînement pour améliorer la stabilité et la vitesse de convergence.
