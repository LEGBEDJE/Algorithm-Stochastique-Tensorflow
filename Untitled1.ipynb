```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
import matplotlib.pyplot as plt
import numpy as np
```

```python
import numpy as np

def metropolis_hastings_pi(N, delta):
    # Initialisation
    x, y = 0.0, 0.0
    count_inside_circle = 0

    for i in range(N):
        # Gnrer un nouveau point candidat
        x_new = x + delta * (np.random.uniform(-1, 1))
        y_new = y + delta * (np.random.uniform(-1, 1))

        # Vrifier si le nouveau point est dans le carr unit
        if abs(x_new) <= 1 and abs(y_new) <= 1:
            # Probabilit d'acceptation
            if x_new**2 + y_new**2 <= 1:
                alpha = 1.0
            else:
                alpha = 0.0

            u = np.random.uniform(0, 1)
            if u < alpha:
                x, y = x_new, y_new

        # Compter les points dans le cercle
        if x**2 + y**2 <= 1:
            count_inside_circle += 1

    # Estimer pi
    pi_estimate = 4 * (count_inside_circle / N)
    return pi_estimate

# Paramtres de l'algorithme
N = 1000000  # Nombre d'itrations
delta = 0.1  # Longueur de pas

# Estimation de pi
pi_estimated = metropolis_hastings_pi(N, delta)
print(f"Estimation de pi: {pi_estimated}")
```

```python
# Charger le dataset MNIST
mnist = tf.keras.datasets.mnist  #est une rfrence au module MNIST dans la bibliothque Keras, qui est incluse dans TensorFlow. Ce module contient des fonctions pour charger le dataset MNIST.
(x_train, y_train), (x_test, y_test) = mnist.load_data() #(x_train, y_train) : Les images et les labels (tiquettes) pour l'entranement.
                                                           #(x_test, y_test) : Les images et les labels pour le test
x_train, x_test = x_train / 255.0, x_test / 255.0  # Normalisation des donnes

# Remodeler les donnes
x_train = x_train.reshape(-1, 784)
x_test = x_test.reshape(-1, 784)
```

    Les images dans x_train et x_test sont initialement reprsentes par des valeurs de pixels allant de 0  255.
    La normalisation est effectue en divisant chaque valeur de pixel par 255.0, ce qui transforme les valeurs des pixels pour qu'elles se situent dans l'intervalle [0, 1].
    La normalisation aide les rseaux de neurones  converger plus rapidement et  amliorer les performances de l'entranement.

    Remodeler les donnes
        reshape(-1, 784) transforme chaque image de 28x28 pixels en un vecteur de 784 lments (28*28 = 784).
    Le -1 permet  NumPy ou TensorFlow de dterminer automatiquement le nombre de samples bas sur la dimension totale et les autres dimensions spcifies. Par exemple, si x_train contient 60 000 images de 28x28, le reshape va transformer x_train en un tableau de dimension (60000, 784).
    Cette transformation est ncessaire parce que la plupart des modles de rseaux de neurones denses (Fully Connected) attendent des vecteurs en entre plutt que des matrices (images).

```python
#construit et retourne un modle de rseau de neurones en utilisant l'API squentielle de Keras.
def create_model():
    model = Sequential([
        Dense(128, activation='relu', input_shape=(784,)), # premire couche de 128 neuronne
        Dense(10, activation='softmax')# deuxime couche de  10 neurones qui correspond au nombre de classes dans le dataset MNIST (les chiffres de 0  9).
    ])
    return model
```

    Sequential est une classe de Keras qui est utilise pour empiler des couches de manire linaire.
    La cration d'un objet Sequential signifie que nous allons ajouter des couches  ce modle, une par une, dans l'ordre.

```python
# Mthode 1 : Batch Gradient Descent

batch_model = create_model()
#Configure le modle avec l'optimiseur Stochastic Gradient Descent (SGD), une fonction de perte adapte  la classification multi-classes
batch_model.compile(optimizer=SGD(learning_rate=0.01), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
#Entrane le modle sur 10 poques
batch_history = batch_model.fit(x_train, y_train, epochs=10, batch_size=len(x_train), validation_data=(x_test, y_test), verbose=0)

# Mthode 2 : Stochastic Gradient Descent
sgd_model = create_model()
sgd_model.compile(optimizer=SGD(learning_rate=0.01), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
sgd_history = sgd_model.fit(x_train, y_train, epochs=10, batch_size=1, validation_data=(x_test, y_test), verbose=0)

# Mthode 3 : Mini-batch Gradient Descent
mini_batch_model = create_model()
mini_batch_model.compile(optimizer=SGD(learning_rate=0.01), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
mini_batch_history = mini_batch_model.fit(x_train, y_train, epochs=10, batch_size=32, validation_data=(x_test, y_test), verbose=0)
```

```python
def plot_history(histories, titles):
    plt.figure(figsize=(12, 8))
    for i, history in enumerate(histories):
        plt.plot(history.history['loss'], label=f'{titles[i]} - train')
        plt.plot(history.history['val_loss'], label=f'{titles[i]} - val')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.title('Comparison of Training Methods')
    plt.show()

# Plot the results
plot_history([batch_history, sgd_history, mini_batch_history], ['Batch Gradient Descent', 'Stochastic Gradient Descent', 'Mini-batch Gradient Descent'])
```

Difficults rencontres

Non-convexit de la fonction de perte : La fonction de perte des rseaux de neurones est gnralement non-convexe et peut contenir de nombreux minima locaux et points de selle, ce qui complique l'optimisation. Le SGD, en raison de sa nature stochastique, peut aider  viter les minima locaux en introduisant du bruit dans le processus de mise  jour des poids, ce qui permet au modle de s'chapper des valles peu profondes.Dans notre cas on a utiliser : 'sparse_categorical_crossentropy cette fonction est utilise pour les problmes de classification multi-classes o les tiquettes sont des entiers (par opposition  categorical_crossentropy o les tiquettes sont encodes en one-hot).
Pendant l'entranement, cette fonction de perte calcule la diffrence entre les prdictions du modle et les tiquettes relles et retourne un score indiquant combien le modle se trompe.

Choix du taux d'apprentissage (learning_rate=0.01): Un taux d'apprentissage trop lev peut entraner une divergence de l'algorithme, tandis qu'un taux trop bas peut ralentir la convergence. Des techniques comme le taux d'apprentissage adaptatif (ex. Adam, RMSprop) ajustent dynamiquement le taux d'apprentissage pendant l'entranement pour amliorer la stabilit et la vitesse de convergence.
